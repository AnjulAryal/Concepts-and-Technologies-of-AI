# Matrix Operations:

*1. Prove A.A−1 = I. *

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);
INV_A = np.linalg.inv(A);
# print(INV_A)
DotAinv_a = np.dot(A,INV_A);
print(DotAinv_a)
print(np.allclose(DotAinv_a, np.eye(2)))

```

*2. Prove AB ̸= BA. *

```python
DotAB = np.dot(A,B);
DotBA = np.dot(B,A);
Array_check = np.array_equal(DotAB,DotBA)
print(Array_check)
```

*3. Prove (AB)T = BTAT. *

```python
TransAB = np.transpose(DotAB)
AtranBtran = np.dot(np.transpose(B),np.transpose(A));
check_array = np.array_equal(TransAB,AtranBtran)
print(check_array)
```

# How Fast is Numpy?

*1. Element-wise Addition:*

```python
import numpy as np;
import time;
start = time.time();
elementA = np.random.randint(1,100,1_000_000);
elementB = np.random.randint(1,100,1_000_000);
addition = elementA + elementB;
end = time.time();
print("Time Taken by Numpy Array Addition is",end-start,"'s");
```

*2.Element-wise Multiplication*

```python
import numpy as np;
import time;
start = time.time();
elementA = np.random.randint(1,100,1_000_000);
elementB = np.random.randint(1,100,1_000_000);
addition = elementA * elementB;
end = time.time();
print("Time Taken by Numpy Array Multiplication is",end-start,"'s");
```

*3. Dot Product*

```python
import numpy as np;
import time;
start = time.time();
elementA = np.random.randint(1,100,1_000_000);
elementB = np.random.randint(1,100,1_000_000);
addition = np.dot(elementA,elementB);
end = time.time();
print("Time Taken by Numpy Dot Product is",end-start,"'s");
```

```python
import numpy as np;
import time;
start = time.time();
elementA = np.random.randint(1,100,(1000,1000));
elementB = np.random.randint(1,100,(1000,1000));
addition = np.dot(elementA,elementB);
end = time.time();
print("Time Taken by Numpy Matrix Multiplication is",end-start,"'s");
```
